# Import libraries

```python
import pandas as pd
import os
```

```python
data_dir = '../data/processed/'
```

# CSV Table for Jumping class

```python
jumping = pd.DataFrame(os.listdir(data_dir + 'jumping/'), columns=['image'])
jumping['image'] = data_dir + 'jumping/' + jumping['image']
jumping['label'] = 0
jumping.head(10)
```

# CSV Table for Ducking class

```python
ducking = pd.DataFrame(os.listdir(data_dir + 'ducking/'), columns=['image'])
ducking['image'] = data_dir + 'ducking/' + ducking['image']
ducking['label'] = 1
ducking.head(10)
```

# CSV Table for Running class

```python
running = pd.DataFrame(os.listdir(data_dir + 'running/'), columns=['image'])
running['image'] = data_dir + 'running/' + running['image']
running['label'] = 2
running.head(10)
```

# Merge into final CSV Table

```python
data = pd.concat((jumping, ducking, running))
data.sample(10)
```

```python
data.to_csv('data.csv')
```
